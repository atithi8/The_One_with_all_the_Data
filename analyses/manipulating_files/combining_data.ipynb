```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Code minipulates the dataframes 

```python
episode=pd.read_csv('../data/episode_info.csv')
wiki_info=pd.read_csv("../data/creators.csv")
rachel=pd.read_csv("../data/Characters/Rachel.csv")
monica=pd.read_csv("../data/Characters/Monica.csv")
phoebe=pd.read_csv("../data/Characters/Phoebe.csv")
ross=pd.read_csv("../data/Characters/Ross.csv")
joey=pd.read_csv("../data/Characters/Joey.csv")
chandler=pd.read_csv("../data/Characters/Chandler.csv")
alls=pd.read_csv("../data/Characters/All.csv")
guests=pd.read_csv("../data/guest_stars.csv")
scenes=pd.read_csv("../data/Characters/scenes.csv")
minors=pd.read_csv("../data/Characters/minors.csv")
memorable_quotes=pd.read_csv("../data/number_quotes_in_episodes.csv")
```

```python
#a function to find when a scene takes place at central perk
def central_perk(i):
    if location.find('centralperk')>-1: return True
    else: return False
```

```python
#a function to find when a scene takes place at Monica's
def monica_ap(i):
    location=str(scenes.location.values[i]).replace('at','').replace(' ','')
    extra=str(scenes.extra_scene_info.values[i]).replace('at','').replace(' ','')

    
    if len(extra)==0:
        if location.find('monica,chandler')>-1: return True
        else: return False
    else: 
        if location.find("monica's")>-1: return True
        elif location.find("monicaandrachel")==0: return True
        elif location.find("monicaandchandler")==0: return True
        elif location.find("monicaandphoebe")==0: return True
        elif location.find("chandlerandmonica")==0: return True
        elif location.find("rachelandmonica")==0: return True
        elif location.find("phoebeandmonica")==0: return True
        else: return False
```

```python
# a function to determine when an episode takes place at joey's apartment
# note the last line takes into account when joey moved out for portions of 4 episodes in season 2

def joeys_ap(i):
    location=str(scenes.location.values[i]).replace('at','').replace(' ','')
    extra=str(scenes.extra_scene_info.values[i]).replace('at','').replace(' ','')

    
    if len(extra)==0:
        if location.find('chandler,joey')>-1: return True
        else: return False
    else: 
        if location.find("chandlerandjoey")==0: return True
        elif location.find("joeyandchandler")==0: return True
        elif location.find("chandler's")==0: return True
        elif location.find("joeyandrachel")==0: return True
        elif location.find("chandlerandkip")==0: return True
        elif location.find("chandlerandeddie")==0: return True
        elif location.find("joeyandjanine")==0: return True
        elif location.find("joey's")>-1 and (scenes.season[i]!=2 or scenes.episode[i]!=16 or scenes.episode[i]!=17 or scenes.episode[i]!=18 or scenes.episode[i]!=19): return True
        else: return False
```

```python
#function for finding when an episode takes place at ross's
def ross_s(i):
    location=str(scenes.location.values[i]).replace('at','').replace(' ','')
    extra=str(scenes.extra_scene_info.values[i]).replace('at','').replace(' ','')

    
    if len(extra)==0:
        if location.find('chandler,joey')>-1: return False
        else: return False
    else: 
        if location.find("rossandrachel")==0: return True
        elif location.find("rachelandross")==0: return True
        elif location.find("ross'")>-1: return True
        else: return False
```

```python
# grabing the scenes that fall under each category
# Rachel's room is used twice, once when it is in Monica's apartment and a second time at Joey's
central_perk=scenes.iloc[[i for i in range(0,len(scenes)) if central_perk(i)==True]]
monicas_apartment=scenes.iloc[[i for i in range(0,len(scenes)) if monica_ap(i)==True or i==1253]]
joey_apartment=scenes.iloc[[i for i in range(0,len(scenes)) if joeys_ap(i)==True or i==2942]]
ross_apartment=scenes.iloc[[i for i in range(0,len(scenes)) if ross_s(i)==True]]
```

```python
#counting the number of lines each person said in a given script
monicas_lines=np.array([len(monica.loc[(monica.season==episode.iloc[x].season) & (monica.episode==episode.iloc[x].episode)]) for x in range(0,len(episode))])
rachel_lines=np.array([len(rachel.loc[(rachel.season==episode.iloc[x].season) & (rachel.episode==episode.iloc[x].episode)]) for x in range(0,len(episode))])
ross_lines=np.array([len(ross.loc[(ross.season==episode.iloc[x].season) & (ross.episode==episode.iloc[x].episode)]) for x in range(0,len(episode))])
phoebe_lines=np.array([len(phoebe.loc[(phoebe.season==episode.iloc[x].season) & (phoebe.episode==episode.iloc[x].episode)]) for x in range(0,len(episode))])
joey_lines=np.array([len(joey.loc[(joey.season==episode.iloc[x].season) & (joey.episode==episode.iloc[x].episode)]) for x in range(0,len(episode))])
chandler_lines=np.array([len(chandler.loc[(chandler.season==episode.iloc[x].season) & (chandler.episode==episode.iloc[x].episode)]) for x in range(0,len(episode))])
all_lines=np.array([len(alls.loc[(alls.season==episode.iloc[x].season) & (alls.episode==episode.iloc[x].episode)]) for x in range(0,len(episode))])
minors_lines=np.array([len(minors.loc[(minors.season==episode.iloc[x].season) & (minors.episode==episode.iloc[x].episode)]) for x in range(0,len(episode))])

#the total number of lines spoken
line_totals= monicas_lines+rachel_lines+ross_lines+phoebe_lines+joey_lines+chandler_lines+minors_lines+all_lines

#counting the number of scenes
scene_total=np.array([len(scenes.loc[(scenes.season==episode.iloc[x].season) & (scenes.episode==episode.iloc[x].episode)]) for x in range(0,len(episode))])
central_perk_count=np.array([len(central_perk.loc[(central_perk.season==episode.iloc[x].season) & (central_perk.episode==episode.iloc[x].episode)]) for x in range(0,len(episode))])
monica_rachel_count=np.array([len(monicas_apartment.loc[(monicas_apartment.season==episode.iloc[x].season) & (monicas_apartment.episode==episode.iloc[x].episode)]) for x in range(0,len(episode))])
joey_chandle_count=np.array([len(joey_apartment.loc[(joey_apartment.season==episode.iloc[x].season) & (joey_apartment.episode==episode.iloc[x].episode)]) for x in range(0,len(episode))])
ross_count=np.array([len(ross_apartment.loc[(ross_apartment.season==episode.iloc[x].season) & (ross_apartment.episode==episode.iloc[x].episode)]) for x in range(0,len(episode))])


#number of guest stars
guest_stars=[0]*len(episode)
for i in range(0, len(guests)):
    guest_season=int(guests.iloc[i].unique_id3.split('-')[0])
    guest_episode=int(guests.iloc[i].unique_id3.split('-')[1])
    position_val=episode.loc[(episode.season==guest_season) & (episode.episode==guest_episode)]
    if len(position_val)>0:
        guest_stars[position_val.index[0]]=guests.iloc[i].num_guests
    else:
        guest_stars[episode.loc[(episode.season==guest_season) & (episode.episode==guest_episode-2)].index[0]]=guests.iloc[i].num_guests



#combining with the wiki info
writer=[]
views=[]
month=[]
for i in range(0, len(wiki_info)):
    writer.append(wiki_info.iloc[i].wri)
    views.append(wiki_info.iloc[i].vie)
    month.append(wiki_info.iloc[i].month)
    #add the extra 0 for when we skip this episode and put in all the old stuff we have
    if len(str(wiki_info.iloc[i].epi))>2:
        writer.append(0)
        views.append(0)
        month.append(0)

        
#wiki info into the df
episode["writer"]=writer
episode["views"]=views
episode["month"]=month


#guest_star info into the df
episode["num_guest_stars"]=guest_stars

#number of "scenes" as a oercetage into the df
episode["central_perk_loc"]=central_perk_count/scene_total
episode["monicas_loc"]=monica_rachel_count/scene_total
episode["chandlers_loc"]=joey_chandle_count/scene_total
episode["ross_loc"]=ross_count/scene_total

#percentage of lines said in script note NaN if we divide by zero cause its a part 1 and part 2
episode["monicas_lines"]=monicas_lines/line_totals
episode["rachel_lines"]=rachel_lines/line_totals
episode["ross_lines"]=ross_lines/line_totals
episode["phoebe_lines"]=phoebe_lines/line_totals
episode["joey_lines"]=joey_lines/line_totals
episode["chandler_lines"]=chandler_lines/line_totals
episode["all_lines"]=all_lines/line_totals




```

```python
#adding in the directors who directed 10 or more episodes
#Kevin Bright (as Kevin S. Bright)
directors_big=['Ben Weiss', 'David Schwimmer', 'Gail Mancuso', 'Gary Halvorson', 'James Burrows', 'Kevin Bright', 'Michael Lembeck', 'Peter Bonerz']

#find if the director for the episode is one of the big directors
dics=[]
for i in episode.director:
    found_one=False
    counter=0
    while not found_one and counter<len(directors_big):
        #kevin bright's name has inconsitencies so this case needs to be taken seperately
        if counter==5:
            if str(i).find(directors_big[counter])==0 and str(i).find(directors_big[3])<0:
                dics.append('Kevin Bright')
                found_one=True
        else:
            if str(i).find(directors_big[counter])==0:
                dics.append(directors_big[counter])
                found_one=True
        counter=counter+1
    if not found_one:
        dics.append('other')

#add results to the df
episode["common_director"]=dics
```

```python
#Getting rid of double episodes.  Part 2's do not have a seperate script so must be combined with part 1's
episode.iloc[[35, 36, 95, 96, 119, 120, 135, 136, 144, 145, 168, 169, 192, 193, 216, 217, 234, 235]]
part1=[35, 95, 119, 135, 144, 168, 192, 216, 234]

#initialize that none of them are a 2 parter
part_1=[0]*(len(episode))
episode['part_1']=part_1

for i in part1:
    #combine both part's guest stars
    temp=str(episode.iloc[i].all_cast).replace("[",'').replace("]",'').split(',')
    for s in str(episode.iloc[i+1].all_cast).replace("[",'').replace("]",'').split(','):
        if s not in temp:
            temp.append(s)
    episode.at[i, 'all_cast']=temp
    
    #combine both part's ratings        
    temp1=episode.iloc[i].Rating
    temp2=episode.iloc[i+1].Rating
    reve1=episode.iloc[i].number_of_reviews
    reve2=episode.iloc[i+1].number_of_reviews
    episode.at[i, 'Rating']=((temp1*reve1)+(temp2*reve2))/(reve1+reve2)
    episode.at[i, 'number_of_reviews']=(reve1+reve2)/2
    episode.at[i, 'part_1']=1
    
    #combine both part's summaries
    episode.at[i, 'short_summary']=episode.at[i, 'short_summary']+" "+episode.at[i+1, 'short_summary']
    
#change the guest stars since these 2 cases and a difference between the part 1 and 2s
episode.at[35,'num_guest_stars']=4
episode.at[95,'num_guest_stars']=3

episode=episode.drop([36, 96, 120, 136, 145, 169, 193, 217, 235]).reset_index(drop=True)
```

```python
#Adding if it was a season premier or a season finale to the df
premiere=['prem']
for i in range(1,len(episode)-1):
    current_season=episode.iloc[i].season
    next_season=episode.iloc[i+1].season
    if current_season<next_season:
        premiere.append('fin')
    elif premiere[len(premiere)-1]=='fin':
        premiere.append('prem')
    else:
        premiere.append('reg')
premiere.append('fin')   

episode['epi_type']=premiere
```

```python
#Adding in the number of catchphrases used 
count_catch=[0]*len(episode)
for i in range(0,len(memorable_quotes)):
    seas_epi=str(memorable_quotes.overall_ep[i])
    new_info=0
    if len(seas_epi)==2:
        newinfo=[int(seas_epi[0]),int(seas_epi[1])]
    elif len(seas_epi)==4:
        newinfo=[int(seas_epi[0:2]),int(seas_epi[2:4])]
    elif seas_epi[0:2]=='10':
        newinfo=[10,int(seas_epi[2:3])]
    else:
        newinfo=[int(seas_epi[0]),int(seas_epi[1:3])]
    catch_phrases=int(memorable_quotes.catchphrase[i])
    
    #where is it in the list
    pos=episode[(episode.season==newinfo[0]) & (episode.episode==newinfo[1])].index.values

    count_catch[pos[0]]=catch_phrases

episode['catchphrases']=count_catch
```

```python
#looking at half a standard deviation above and below the mean review to rank as below, average, or above
average=[np.mean(episode.Rating)-1/2*np.std(episode.Rating),np.mean(episode.Rating)+1/2*np.std(episode.Rating) ]
above_average=[np.mean(episode.Rating)+1/2*np.std(episode.Rating),np.max(episode.Rating)]
below_average=[np.min(episode.Rating),np.mean(episode.Rating)-1/2*np.std(episode.Rating) ]

#print the intervals
print(above_average)
print(average)
print(below_average)

#print the number of episodes in each
print(len([1 for i in episode.Rating if average[0]<=i<=average[1] ]))
print(len([1 for i in episode.Rating if above_average[0]<=i<=above_average[1] ]))
print(len([1 for i in episode.Rating if below_average[0]<=i<=below_average[1] ]))

rats=[]
for i in episode.Rating:
    if average[0]<=i<=average[1]: rats.append("av")
    elif above_average[0]<=i<=above_average[1]: rats.append("above")
    elif below_average[0]<=i<=below_average[1]: rats.append("below")

#add to the df
episode["rat_grouped"]=rats
```

```python
##adding in the common name to the title 
about=[]
main=['rachel', 'ross', 'phoebe', 'chandler', 'monica', 'joey']
for title in episode.title:
    newtitle=title.lower()
    setall=[]
    for person in main:
        if newtitle.find(person)>-1:
            setall.append(person)
    if len(setall)==0:
        about.append('other')
    else:
        about.append(setall)

#adding if a main character's name is in the title to the df 
episode["title_about"]=about
```

```python
#my fav reuccuring characters
gunther=[0]*len(episode)
janice=[0]*len(episode)
richard=[0]*len(episode)
heckles=[0]*len(episode)
frank=[0]*len(episode)
estelle=[0]*len(episode)
carol=[0]*len(episode)
for i in range(0,len(episode)):
    cast=episode.iloc[i].all_cast
    if 'James Michael Tyler' in episode.iloc[i].all_cast:
        gunther[i]=1
    if 'Maggie Wheeler' in episode.iloc[i].all_cast:
        janice[i]=1
    if 'Tom Selleck' in episode.iloc[i].all_cast:
        richard[i]=1
    if 'Larry Hankin' in episode.iloc[i].all_cast:
        heckles[i]=1
    if 'Jane Sibbett' in episode.iloc[i].all_cast:
        carol[i]=1
    if 'Anita Barone' in episode.iloc[i].all_cast:
        carol[i]=1
    if 'Giovanni Ribisi' in episode.iloc[i].all_cast:
        #not the condom guy one
        if episode.iloc[i].season != 2 or episode.iloc[i].episode !=6:
            frank[i]=1
    if 'June Gable' in episode.iloc[i].all_cast:
        #not the nurse one
        if episode.iloc[i].season != 1 or episode.iloc[i].episode !=23:
            estelle[i]=1
    
#can't forget ursual but obviously can't use the actor route here
ursula=[0]*len(episode)
for x in range(0,len(episode)):
    if len(minors.loc[(minors.season==episode.iloc[x].season) & (minors.episode==episode.iloc[x].episode) & (minors.character=='ursula')])>0:
        ursula[x]=1
        

#Add the dataFrames in         
episode['gunther']=gunther
episode['janice']=janice
episode['richard']=richard
episode['heckles']=heckles
episode['frank']=frank
episode['estelle']=estelle
episode['ursula']=ursula
episode['carol']=carol
```

```python
#reorder the dataframe into a cleaner way
allofthem=[col for col in episode]
list1=allofthem[0:12]
list2=allofthem[13:24]
list2.append(allofthem[12])
list2.append(allofthem[24])
list2.append(allofthem[26])
list2.append(allofthem[27])

list3=allofthem[29:37]
list3.append(allofthem[25])
list3.append(allofthem[28])

finallist=list1+list2+list3

epi_copy=episode[finallist]
```
