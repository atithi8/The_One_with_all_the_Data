```python
# Our common data handling package
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
```

```python
#The first webite we need to pull in 
orig="https://www.imdb.com/title/tt0583459"
html = urlopen(orig)
soup = BeautifulSoup(html,"html.parser")

#the info we want to get
title=[]
seas=[]
epi=[]
number_of_review=[]
out_of_10=[]
director=[]
release_date=[]
all_cast=[]
next_episode_string=''
plot=[]

while next_episode_string!='Done':
    title.append(soup.head.title.text.replace('"Friends"','')[0:soup.head.title.text.replace('"Friends"','').find('(')])
    season_episode=soup.find('div',{'class':"bp_heading"}).text.split('|')
    
    #progress bar
    print(season_episode)
    
    seas.append(int(season_episode[0].replace('Season ','')))
    epi.append(int(season_episode[1].replace('Episode ','')))
    number_of_review.append(int(soup.find('span',{'itemprop':"ratingCount"}).text.replace(',','')))
    out_of_10.append(float(soup.find('div',{'class':"ratingValue"}).text.replace('\n','').replace('/10','')))
    director.append(soup.find('div',{'class':"credit_summary_item"}).text.replace('\n','').split(':')[1])
    release_date.append(soup.find('a',{'title':"See more release dates"}).text.replace('Episode aired','').replace('\n',''))
    next_episode_string=soup.findAll('a',{'class':"bp_item np_next"})
    plot.append(soup.find('div',{'class':"summary_text"}).text.replace('\n','').strip())
    
    html=urlopen(orig+'/fullcredits')
    soup = BeautifulSoup(html,"html.parser")
    
    all_cast.append([a.text.replace('\n','').replace('...',',').split(',')[0].strip() for a in soup.find_all('tr',{'class',"even"})]+[a.text.replace('\n','').replace('...',',').split(',')[0].strip() for a in soup.find_all('tr',{'class',"odd"})])
    
    #move on to th enext episode and repeat, if not you are done
    if len(next_episode_string)==0:
        next_episode_string='Done'
    else:
        orig='https://www.imdb.com'+next_episode_string[0].get('href')
        html=urlopen(orig)
        soup = BeautifulSoup(html,"html.parser")
        
pd.DataFrame({'season': seas, "episode":epi, "title" : title, 'director': director, 'release_date': release_date,'all_cast': all_cast,'Rating': out_of_10, 'number_of_reviews': number_of_review,'short_summary':  plot }).to_csv('../data/episode_info.csv',index=False)
```
