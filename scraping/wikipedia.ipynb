# Scrape episode info from Wikipedia

```python
### season 1

friends01 = urlopen('https://en.wikipedia.org/wiki/Friends_(season_1)')

soup01 = BeautifulSoup(friends01, 'html.parser')
```

```python
# select the interest part of the page
soup_table = soup01.find('table', {'class':'wikitable plainrowheaders wikiepisodetable'})

soup_tr = soup_table.find_all('tr',{'class':'vevent'})
```

```python
# check we are getting the correct information
for td in soup_tr[0].find_all('td'):
    print(td.text)
```

```python
# create a dictionary to store all the scraped information

df_dict = {'epi':[],'ttl':[],'dir':[],'wri':[],   # epi = episode, ttl = title, dir = director, wri = writers
          'air':[],'code':[],'vie':[]}            # air = airdate, code = production code, vie = no. US viewers in M

df_dict_keys = ['epi','ttl','dir','wri','air','code','vie']

# scrape the information and store it in the dictionary
for row in soup_tr:
    tds = row.find_all('td')
    for i in range(len(tds)):
        df_dict[df_dict_keys[i]].append(tds[i].text)
```

```python
# convert the dictionary into a dataframe and add season column
friends01 = pd.DataFrame(df_dict)
friends01['season'] = 1

friends01
```

### Create dataframe

```python
# Create the dataframe with all seasons
frames = [friends01, friends02, friends03, friends04, friends05,
         friends06, friends07, friends08, friends09, friends10]

misc = pd.concat(frames, ignore_index = True)
```

```python
# Clean string variables
# Create air month column
misc['vie'] = misc['vie'].str.replace(r'\[citation needed\]', '')
misc['vie'] = misc['vie'].str.replace(r'\[1\]', '')
misc['vie'] = misc['vie'].str.replace(r'\[2\]', '')
misc['vie'] = misc['vie'].str.replace(r'\[3\]', '')
misc['air'] = misc['air'].str.replace(r'\[a\]', '')
misc['air'] = misc['air'].str.replace(r'\)', '')
misc['air'] = misc['air'].str[-10:]
misc['month'] = misc['air'].str[5:7]
```

```python
misc['wri'] = misc['wri'].astype(str).str.replace('Story by \ : ','')
misc['wri'] = misc['wri'].astype(str).str.replace('Teleplay by \ :','&')
misc['wri'] = misc['wri'].astype(str).str.replace('\n','')
misc['ttl'] = misc['ttl'].astype(str).str.replace('"','')
```

```python
# Save final df
misc.to_csv('../data/creators.csv' , index = False) # Path needs updating to new repo
```

# Scrape episodes summaries

```python
### season 1

friends01 = urlopen('https://en.wikipedia.org/wiki/Friends_(season_1)')

soup01 = BeautifulSoup(friends01, 'html.parser')

soup_table = soup01.find('table', {'class':'wikitable plainrowheaders wikiepisodetable'})

soup_sum = soup_table.find_all('tr',{'class':'expand-child'})

df_dict = {'summary':[]}            

df_dict_keys = ['summary']

for row in soup_sum:
    tds = row.find_all('td')
    for i in range(len(tds)):
        df_dict[df_dict_keys[i]].append(tds[i].text)
        
summ01 = pd.DataFrame(df_dict)

summ01['episode'] = summ01.index
for i in summ01.episode:
    summ01.episode[i] = i+1

summ01['season'] = 1

summ01
```
